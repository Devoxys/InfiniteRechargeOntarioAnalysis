```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline


def scatterplot(df, tx, ty):
    """Generates scatterplot with line of best fit"""
    title = ty+" vs. "+tx
    plot = df.plot(kind='scatter', figsize=(10, 5), title=title, x=tx, y=ty)
    plot.set_xlabel(tx)
    plot.set_ylabel(ty)
    
    # Uncomment the two lines below and make appropriate edits to label a particular team in a graph
    #team_lbl = df.loc[df['Team'] == 'frc4308']
    #plot.annotate('frc4308', (team_lbl[tx], team_lbl[ty]), textcoords='offset points', xytext=(0, 5), ha='center')
    # Make sure the team is in the dataframe/csv, of course
    
    x = df[tx]
    y = df[ty]
    bm = np.polynomial.polynomial.Polynomial.fit(x, y, 1)
    b, m = bm.convert().coef
    linestr = 'y = '+str(m)+'x + '+str(b)
    plot.plot(x, m*x+b)
    corr_mat = np.corrcoef(x, y)
    r2 = (corr_mat[0, 1])**2
    
    print(linestr)
    print("R^2 = " + str(r2))
    
    
    
```

# What correlates to ranking high?

The goal of teams in the district format is to perform well in events to rank high, based on the amount of district points earned. A certain amount of teams qualify for District Championships (in 2020ont it is about top 80) where more points are earned and the top 30 teams qualify for World Championships. District Points are the sum of points awarded based on performance during qualification and playoff at events, placement in alliance selection, and awards. However, those aren't the only measures of performance in FRC. Interesting questions to ask are: how important are each of the district point categories, is there are particular aspect of a game that indicates how strong a team is or is focused on by high ranking teams, how powerful is a qualification win or being picked early, etc. 

```python
df = pd.read_csv('2020ont_team_data.csv')

# Making a ranked dataframe. 
ranked_df = df.sort_values(by=['District Points'], ascending=False)
ranked_df = ranked_df.reset_index()
ranked_df['Rank'] = ranked_df['District Points'].rank(method='max', ascending=False) 
# While FRC uses tiebreakers, they are not implemented at this time
ranked_df
```


---

## Points

```python
scatterplot(ranked_df, 'Rank', 'District Points')
```

Rank is based on District Points. The graph is clean and gives you a sense of distribution.

```python
scatterplot(ranked_df, 'Rank', 'Qualification Points')
```

```python
scatterplot(ranked_df, 'Rank', 'Alliance Points')
```

Award and Elimination Points do not yield interesting graphs (essentially everything is split between a couple levels) for obvious reasons and I will refer you to the rulebook for more. These twohowever are worth taking a look at.

Qualification Points yield slightly more than Alliance Points on average (at 30 maximum, Elimination awards the most on its own). However, the nature of Alliance selection process make it a tighter indicator of rank. It appears that teams become captain or get picked by other teams close to where they perform relatively. In Qualification on the other hand, strong teams still tend to come out on top more often than not but random teammates and random matchups clearly affect how things turn out. Bad synergy or tough matchups can make it seem like you're underperforming.

---

## Qualification

```python
scatterplot(ranked_df, 'Rank', 'Qualification Win %')
```

The percentage of qualification matches a team wins only seems mildly correlated with ranking high. It does appear that the strongest of teams vastly overperform. This could be a sign that those very few teams have the potential to win matches pretty much on their own prowess during qualifications. 

```python
scatterplot(ranked_df, 'Rank', 'Top Qualification Rank')
```

What place a team ends up after qualifications does have a stronger correlation than the number of matches won. A notable observation is that it appears that throughout the graphs shown thus far, there are usually more points that fall on the "positive" side of the line of best fit but those that fall on the "negative side" tend to be far more negative. In this graph where "positive" is closer to 0 on the y-axis, this continues to be true. Outliers tend to be of the not-so-good kind.

```python
scatterplot(ranked_df, 'Rank', 'Top Ranking Score')
```

A team's Ranking Score is the number of Ranking Points (earned during quals) divided by the games they've played. The correlation strength is unsurprisingly similar the qualification rank, after all, that's how its determined, but it shares some notable characteristics with the Qualification Win % graph. Notice the group of strong overperformers at the very left of the graph in both this and the Win % graph.

---

## Alliance Selection and Playoffs

```python
unseeded = ranked_df['Top Alliance Seed'] == 0.0
ranked_df.loc[unseeded, 'Top Alliance Seed'] = 9.0 # Move unseeded to the bottom
scatterplot(ranked_df, 'Rank', 'Top Alliance Seed')
```

You can see some of the trends in Alliance Selection in this graph. The top seeds pick teams that are at a similar rank (in event and in district) and pick a low to medium ranked team when they eventually get their second pick. Being picked by those teams could represent a significant boost to district ranking because of the amount of points to be had for placing high in playoffs, which could explain why there are no teams below top 40 that have been part of a the first seeded alliance. On the other hand, this trend could condemn lower seeded teams to lower ranks (even if they are team captain or first pick). 

```python
no_level = ranked_df['Playoff Level'] == 0.0
ranked_df.loc[no_level, 'Playoff Level'] = 4.0 # Moving non-playoff teams
scatterplot(ranked_df, 'Rank', 'Playoff Level')
```

Notably the trendline for this 4-tiered graph has had the strongest correlation thus far. The ranks appear to be split between sections based on where teams ended up in the playoffs. This is due to the circumstances of this season and may or many not be an accurate representation of how things will end up in a complete season. Comparing this graph and the one before is shows evidence that seeding is quite accurate to actual placement in the playoffs.

## Game

Reminder: the "cumulative" in the names of these categories refer to how it is the sum of the points earned in the specific sections of the games that the team played in. It includes contribution by all other teams on the alliance. This is how the data is provided by APIs because FRC does not track individual team statistics. That is the purpose of scouting during events.

```python
ranked_df['Top Game Points'] = ranked_df['Top Cumulative Auto'] + ranked_df['Top Cumulative Endgame'] + ranked_df['Top Cumulative Teleop']
scatterplot(ranked_df, 'Rank', 'Top Game Points')
```

While there clearly is a trend - top teams get lots more points than bottom ranked teams - the relatively low R^2 appears to be because of the distribution around the top of the rankings. I think the outliers (the 2 teams at around top 10 with vastly more points than the rest and the teams around top 20 with relatively low amounts of points) are worth investigating. This excludes points earned through penalties so super high scoring teams must have been through a couple of super high scoring "unicorn" games.

```python
scatterplot(ranked_df, 'Rank', 'Top Cumulative Auto')
```

There appears to be a trend but there everything is all over the place. Amongst top, middle and bottom ranked teams, there are teams that played games with lots of points earned in auto, an average amount, or a low amount of points earned in auto. 

```python
scatterplot(ranked_df, 'Rank', 'Top Cumulative Endgame')
```

Even looser than the previous. I suspect that this is because of how easy to it to earn some points in these phases (crossing the line or parking in the zone) makes it easy for teams to earn a decent amount of points here even passisvely. 

```python
scatterplot(ranked_df, 'Rank', 'Top Cumulative Teleop')
```

Here the R^2 isn't above 0.5 but the graph shows a trend, with the correlation strength likely being thrown off a little by the the big positive outliers. Overall however, there seem to be a lot more teams close to an average level in teleop. 


Looking at the scales, the points earned in teleop is more spread out than auto and endgame. All 3 phases provide account for relatively the same amount of points. The data shown here does not provide any conclusion on the best phase to focus in in the game. In my opinion, teleop having the strongest correlation and tighter distribution could make it worth pushing forward on if you want to make it look like you stand out (though, an amazing climb or a 5-9 ball auto is sure to do that as well). I believe this topic deserves more exploration and I likely will cover it in a different part and notebook.

---

With linear trendlines the strongest correlations with rank are with Playoff Level and Alliance Points with R^2>0.7 followed by a number of other indicators with R^2>0.6. Generally, **indicators used in FRC to determine/indicate some sort of ranking** are strongest indicators of district ranking outcomes. Qualification and Alliance Points contribute directly to District Ranking, Alliance Selection related indicators like Alliance Points and Alliance Seed are particularly strong, so are Top Qualification Rank and its base Ranking Score. Getting to a high place in playoffs is already seen as a good indicator of teams that will show up in District Championships and even World Championships, the data confirms this.
